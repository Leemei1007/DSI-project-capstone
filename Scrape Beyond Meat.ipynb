```python
import requests
from requests_html import HTMLSession
from bs4 import BeautifulSoup
import pandas as pd

import time
```

Source codes: https://www.youtube.com/c/JohnWatsonRooney/videos

```python
def flatten(t):
    return [item for sublist in t for item in sublist]
```

```python
# Beyond Meat Burger

def get_data(page):
    r = requests.get(f'https://www.amazon.sg/Beyond-Meat-burger-227g-Frozen/product-reviews/B074K5MDLW/ref=cm_cr_getr_d_paging_btm_next_13?ie=UTF8&reviewerType=all_reviews&pageNumber={page}')
    sp = BeautifulSoup(r.text, 'html.parser')
    return sp
```

```python
def parse(sp):
    res = []
    products = sp.find_all('div', {'data-hook': 'review'})
    for item in products:
        try:
            reviews = {
                'title': item.find('span', {'data-hook': 'review-title'}).text,
                'body': item.find('span', {'data-hook': 'review-body'}).text,
                'rating': item.find('i', {'data-hook': 'cmps-review-star-rating'}).text,
                'reviewer_name': item.find('span', {'class': 'a-profile-name'}).text,
            }
            print(reviews)
            
            res.append(reviews)
        except:
            pass
    return res 

totals = []

for x in range(1,14):
    time.sleep(5)
    print('getting page', x)
    html = get_data(x)
    totals.append(parse(html))
    print()
    if not html.find('div', {'id': 'cm_cr-pagination_bar'}).find('li', {'class': 'a-last'}):
        break

df = pd.DataFrame(flatten(totals))
df.to_csv('BeyondMeatBurger-results.csv', index = False)
```
