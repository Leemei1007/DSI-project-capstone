```python
def sentiment(x):
    '''
    Input will be a string
    Output returns the sentiment score of the string
    '''
    sent = SentimentIntensityAnalyzer() # Instantiate
    return sent.polarity_scores(x) # Calculating the scores of each document
```

```python
MFcombine['sentiment'] = MFcombine['review'].apply(sentiment)
```

```python
MFcombine.head(2)
```

```python
MFcombine['word_compound']  = MFcombine['sentiment'].apply(lambda score_dict: score_dict['compound'])
```

```python
MFcombine.head(1)
```

```python
def assign_compound(x):
    if x == 0:
        return 3
    elif x > 0:
        return 5
    elif x < 0:
        return 1
```

```python
MFcombine['word_comp_score'] = MFcombine['word_compound'].apply(assign_compound)
```

```python
MFcombine.head(5)
```

```python
MFcombine[['ratings', 'word_compound', 'word_comp_score']].head()
```

```python
MFcombine['difference'] = MFcombine['ratings'] - MFcombine['word_comp_score']
```

```python
MFcombine['status'] = np.where(MFcombine['difference']<=2, 'correct', 'incorrect' )
```

```python
MFcombine['status'].value_counts()
```

```python
MFcombine.head(1)
```

```python
# set up x and y

X = MFcombine['review']
y = MFcombine['word_compound']
```

```python
# train test split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.4)
```

```python
tf = TfidfVectorizer()
```

```python
X_train_tf = tf.fit_transform(X_train).todense()
```

```python
X_test_tf = tf.transform(X_test).todense()
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train_tf, y_train)
```

```python
y_pred = lr.predict(X_test_tf)
```

```python
print ('RMSE is: \n', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print ('R-squared is: \n', round(r2_score(y_test, y_pred),3))
```

```python
X_train_tf = pd.DataFrame(tf.transform(X_train).todense(), columns = tf.get_feature_names())
X_test_tf = pd.DataFrame(tf.transform(X_test).todense(), columns = tf.get_feature_names())
```

```python
explainer = shap.LinearExplainer(lr, X_train_tf, feature_perturbation="interventional")
shap_values = explainer.shap_values(X_train_tf)
```

```python
shap.summary_plot(shap_values, X_train_tf, feature_names=tf.get_feature_names())
```
